# Decision Tree Classifier 

For the given ‘Iris’ dataset, create the Decision Tree classifier and visualize it graphically.
The purpose is if we feed any new data to this classifier, it would be able to predict the right class accordingly.

```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
dataset = pd.read_csv('Iris.csv')
```

```python
dataset.head(5)
```

```python
dataset.shape
```

```python
dataset['Species'].value_counts()
```

```python
x=dataset.iloc[:,1:5].values
y=dataset['Species']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3, random_state=0)
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(criterion='entropy')
clf.fit(x_train,y_train)

y_pred = clf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
print('Accuracy: ', accuracy_score(y_test,y_pred).round(4)*100)
```

```python
from sklearn.metrics import confusion_matrix
c = confusion_matrix(y_test,y_pred)
print('Confusion Matrix: \n ', c)
```

```python
print('Heatmap of confusion matrix')
plt.figure(figsize=(4,4))
ax=sns.heatmap(c,annot= True, cmap='Blues', cbar=False)
bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5)
```

```python
from sklearn import tree
plt.figure(figsize=(18,10))
tree.plot_tree(clf.fit(x,y),feature_names=dataset.columns, filled=True,
               precision=3, proportion=True,rounded=True)
plt.show()
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
clf1 = KNeighborsClassifier(n_neighbors = 5 , metric = 'minkowski', p=2)
clf1.fit(x_train,y_train)

y_pred1 = clf1.predict(x_test)
```

```python
print('Accuracy: ', accuracy_score(y_test,y_pred1).round(4)*100)
```

```python
from sklearn.metrics import confusion_matrix
c = confusion_matrix(y_test,y_pred1)
print('Confusion Matrix: \n ', c)
```

# Support Vector Machine

```python
from sklearn.svm import SVC
clf2 = SVC(kernel = 'linear',random_state =0,gamma='auto')
clf2.fit(x_train,y_train)

y_pred2 = clf2.predict(x_test)
```

```python
print('Accuracy: ', accuracy_score(y_test,y_pred2).round(4)*100)
```

```python
from sklearn.metrics import confusion_matrix
c = confusion_matrix(y_test,y_pred2)
print('Confusion Matrix: \n ', c)
```

# Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
clf3 = GaussianNB()
clf3.fit(x_train,y_train)

y_pred3 = clf3.predict(x_test)
```

```python
print('Accuracy: ', accuracy_score(y_test,y_pred3).round(4)*100)
```

```python
from sklearn.metrics import confusion_matrix
c = confusion_matrix(y_test,y_pred3)
print('Confusion Matrix: \n ', c)
```
